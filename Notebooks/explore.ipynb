```python
if "play_time_ms" in df.columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df["play_time_ms"], bins=50, kde=True)
    plt.title("Distribution of Play Time (ms)")
    plt.xlabel("Play Time (ms)")
    plt.show()
```

```python
if "is_click" in df.columns:
    plt.figure(figsize=(5,3))
    sns.countplot(x="is_click", data=df)
    plt.title("Click vs Non-Click")
    plt.show()
```

```python
if "user_id" in df.columns and "date" in df.columns:
    df_sorted = df.sort_values(by=["user_id", "date"])
    print("First few rows of sorted data by user and date:")
    print(df_sorted.head()) 
```

```python
user_sequences = df_sorted.groupby("user_id", group_keys=False).apply(
    lambda x: x.drop(columns=["user_id"]).to_dict(orient="records")
)
```
